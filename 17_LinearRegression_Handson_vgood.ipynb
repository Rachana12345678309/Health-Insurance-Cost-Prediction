```python
df['children'] = df['children'].astype('object') #since children should not be a decimal value during calculations
```

```python
#numerical columns - fill the null values with mean
#categorical columns - fill the null values with mode
```

```python
for i in col:
  if df[i].dtype == 'object':
    df[i].fillna(df[i].mode()[0],inplace=True) #[0] fetches the value at index 0, or the top value..most occuring value
  else:
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# Duplicate values
```

```python
df.duplicated().sum() #if exists, df.drop_duplicates(inplace = True)
```

```python
# Outlier values
```

```python
for i in col:
  if (df[i].dtype != 'object'): #numerical columns only will contain outliers
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
# Removing outliers

for i in col:
  if df[i].dtype != 'object':
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3-q1
    lower_limit = q1 - 1.5*iqr
    upper_limit = q3 + 1.5*iqr
    df = df[(df[i] >= lower_limit) & (df[i] <= upper_limit)]
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
encoders = {}
mappings = {}
for i in df.columns:
  if df[i].dtype == 'object':
    le = LabelEncoder() # new instance every time
    df[i] = le.fit_transform(df[i])
    encoders[i] = le # storing the correct encoder for each column
    mappings[i] = dict(zip(le.classes_,le.transform(le.classes_)))
```

```python
#linear regression

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error,mean_absolute_percentage_error
```

```python
#problem statement was about - charge prediction for each customer based on their existing conditions hence charge to be taken as target column
X = df.drop(columns = ['charges']) #all independent cols (i.e other than charges column)
y = df['charges'] #dependent col
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
my_model = LinearRegression()
```

```python
#training the model
my_model.fit(X_train,y_train)
```

```python
#testing
y_pred = my_model.predict(X_test)
```

```python
y_pred
```

```python
result = pd.DataFrame(columns = ['Actual_value', 'Predicted_value'])
result['Actual_value'] = y_test
result['Predicted_value'] = y_pred
```

```python
result
```

```python
result = r2_score(y_test,y_pred)
```

```python
result #96% of the model is showing the correct results
```

```python
mae = mean_absolute_error(y_test,y_pred)
print(mae)
mse = mean_squared_error(y_test,y_pred) #will be a big number
print(mse)
rmse = np.sqrt(mse)
print(rmse)
mape = mean_absolute_percentage_error(y_test,y_pred)
print(mape)
```

```python
#The model's avg error is 8.98% of actual values.excellent

usually
< 10 % -> excellent

10-20% -> Good

20-50% -> Acceptable

>50% -> poor
```
